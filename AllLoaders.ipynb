# JSON Loader

```python
from langchain_community.document_loaders import JSONLoader
```

```python
basepath = "/content/drive/MyDrive/Langchain/"
```

```python
import os
file_path = os.path.join(basepath,"facebook_chat.json")
```

```python
!pip install jq
```

```python
loader = JSONLoader(
    file_path=file_path,
    jq_schema=".messages[].content",
    text_content=False
)
```

```python
docs = loader.load()
print(len(docs))
print(docs[0].page_content)
```

```python
import json
with open(file_path) as f:
  data = json.load(f)
pprint.pprint(data)
```
